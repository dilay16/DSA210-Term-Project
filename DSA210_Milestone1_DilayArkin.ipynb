```python
import matplotlib.pyplot as plt

# Histogram
df['ai_usage_minutes'].hist()
plt.title("AI Usage Distribution")
plt.tight_layout()
plt.show()

# Scatter
plt.scatter(df['ai_usage_minutes'], df['productivity'])
plt.xlabel("AI Usage")
plt.ylabel("Productivity")
plt.title("AI Usage vs Productivity")
plt.tight_layout()
plt.show()

# Bar chart
df['purpose'].value_counts().plot(kind='bar')
plt.title("Purpose of AI Use")
plt.tight_layout()
plt.show()

# Boxplot
df.boxplot(column='ai_usage_minutes', by='course_type')
plt.title("AI Usage by Course Type")
plt.suptitle("")
plt.tight_layout()
plt.show()
```

```python
# Correlation
df[['ai_usage_minutes', 'productivity']].corr()
```

```python
# T-test
from scipy.stats import ttest_ind

quant = df[df['course_type'] == 'quantitative']['ai_usage_minutes']
qual = df[df['course_type'] == 'qualitative']['ai_usage_minutes']

ttest_ind(quant, qual)
```

**EDA + ANALYSIS**

The dataset shows variation in AI usage across different days and contexts. Most AI usage values range between approximately 25 and 120 minutes.

The correlation analysis indicates a weak negative relationship between AI usage and productivity. This suggests that higher AI usage may be slightly associated with lower productivity, but the relationship is not strong.

This implies that AI usage alone does not fully explain productivity levels, and other factors such as task type, difficulty, and deadlines may also play a significant role.

The t-test comparing AI usage between quantitative and qualitative courses resulted in a statistically significant difference (p < 0.05), indicating that AI usage differs across course types.

However, given the variability in the dataset, these results should be interpreted cautiously.

**HYPOTHESIS TESTING**

The correlation between AI usage and productivity was found to be weak and negative (approximately -0.19). This indicates that higher AI usage may be slightly associated with lower productivity, but the relationship is not strong.

The t-test comparing AI usage between quantitative and qualitative courses resulted in a p-value significantly below 0.05, indicating a statistically significant difference between the two groups.

Therefore, the null hypothesis is rejected, and it can be concluded that AI usage differs significantly between quantitative and qualitative courses.

However, given the variability in the dataset and the relatively weak correlation, these results should be interpreted with caution.
