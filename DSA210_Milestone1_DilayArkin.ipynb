```python
import matplotlib.pyplot as plt

# Histogram
df['ai_usage_minutes'].hist()
plt.title("AI Usage Distribution")
plt.tight_layout()
plt.show()

# Scatter
plt.scatter(df['ai_usage_minutes'], df['productivity'])
plt.xlabel("AI Usage")
plt.ylabel("Productivity")
plt.title("AI Usage vs Productivity")
plt.tight_layout()
plt.show()

# Bar chart
df['purpose'].value_counts().plot(kind='bar')
plt.title("Purpose of AI Use")
plt.tight_layout()
plt.show()

# Boxplot
df.boxplot(column='ai_usage_minutes', by='course_type')
plt.title("AI Usage by Course Type")
plt.suptitle("")
plt.tight_layout()
plt.show()
```

```python
# Correlation
df[['ai_usage_minutes', 'productivity']].corr()
```

```python
# T-test
from scipy.stats import ttest_ind

quant = df[df['course_type'] == 'quantitative']['ai_usage_minutes']
qual = df[df['course_type'] == 'qualitative']['ai_usage_minutes']

ttest_ind(quant, qual)
```

**EDA + ANALYSIS**

The dataset shows variation in AI usage across different days and contexts. Most AI usage values range between approximately 25 and 120 minutes, with higher values observed on days with deadlines.

The scatter plot suggests a clear negative relationship between AI usage and productivity. Higher AI usage levels are generally associated with lower productivity values, especially on days with deadlines.

The boxplot indicates that AI usage is higher in quantitative courses compared to qualitative courses, suggesting that students rely more on AI tools in quantitatively demanding tasks.

**HYPOTHESIS TESTING**

The correlation between AI usage and productivity was found to be strong and negative (-0.94). This indicates that higher AI usage is associated with lower productivity in this dataset.

The t-test comparing AI usage between quantitative and qualitative courses resulted in a p-value of 0.0011, which is below the significance level of 0.05. This indicates that there is a statistically significant difference in AI usage between the two groups.

Therefore, the null hypothesis is rejected, and it can be concluded that AI usage differs significantly between quantitative and qualitative courses.
